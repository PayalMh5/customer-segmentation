```python
x = df.iloc[:,2: ].values
x
```

```python
scale = MinMaxScaler() 
features = scale.fit_transform(x) 
features
```

```python
inertia = []
for i in range(1, 16):
    k_means = KMeans(n_clusters=i, n_init=10)  # Explicitly set n_init
    k_means = k_means.fit(features)
    inertia.append(k_means.inertia_)
```

```python
# Plot the elbow graph
plt.figure(figsize=(20, 7))

# Elbow graph
plt.subplot(1, 2, 1)
plt.plot(range(1, 16), inertia, 'bo-')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Graph')
```

```python
# KElbowVisualizer
plt.subplot(1, 2, 2)
kmeans = KMeans(n_init=10)  # Explicitly set n_init
visualize = KElbowVisualizer(kmeans, k=(1, 16))
visualize.fit(features)
visualize.finalize()

plt.suptitle("Elbow Graph and Elbow Visualizer")
plt.show()
```

```python
silhouette_avg = []

for i in range(2, 16):  # Start from 2 because silhouette score is not defined for a single cluster
    kmeans = KMeans(n_clusters=i)  # Explicitly set n_init and random_state for reproducibility
    cluster_labels = kmeans.fit_predict(features)
    silhouette_avg.append(silhouette_score(features, cluster_labels))
```

```python
# Plotting the silhouette scores
plt.figure(figsize=(10, 5))
plt.plot(range(2, 16), silhouette_avg, 'bo-')
plt.xlabel('Number of clusters')
plt.ylabel('Average Silhouette Score')
plt.title('Silhouette Analysis for KMeans Clustering')
plt.show()
```

```python
# Plotting the silhouette scores
plt.figure(figsize=(10, 7))
plt.plot(range(2, 16), silhouette_avg, 'bX-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette score')
plt.title('Silhouette Analysis for Optimal k')
plt.show()
```

```python
model = KMeans(n_clusters=3, n_init=10) 
model = model.fit(features)
```

```python
y_km = model.predict(features) 
centers = model.cluster_centers_
```

```python
df['Cluster'] = pd.DataFrame(y_km) 
df.to_csv("Cluster_data",index=False)
```
